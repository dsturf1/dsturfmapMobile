```python
target_courseName = '오너스'
```

```python
files = glob.glob(os.path.join(json_folder,'*.geojson'))
files = [x for x in files if target_courseName in x]
files
```

```python
whole_df = gpd.GeoDataFrame([])
import uuid

for file_ in files:
  gdf = gpd.read_file(file_)
  whole_df = pd.concat([whole_df, gdf])
whole_df = whole_df.reset_index(drop = True)
whole_df['Id'] = [uuid.uuid4() for _ in range(len(whole_df.index))]
whole_df['Id'] = whole_df['Id'].astype(str)
```

```python
whole_df
```

```python
files = glob.glob(os.path.join(json_folder,'*.json'))
file_ = [x for x in files if 'area' in x][0]
```

```python
area_df = pd.read_json(file_, orient='records')
```

```python
files = glob.glob(os.path.join(json_folder,'*.json'))
file_ = [x for x in files if 'course' in x][0]
course_df = pd.read_json(file_, orient='records')
```

```python
course_id = course_df.loc[course_df.name.str.contains(target_courseName),'id'].item()
course_id
```

```python
whole_df = whole_df.merge(area_df, left_on = 'Type', right_on = 'name', how = 'left')
```

```python

whole_df = whole_df[['Id','Name', 'Client', 'Course', 'Hole', 'Type', 'Valid',
       'TypeId', 'color', 'display', 'area_def', 'work_def',
       'DSZindex', 'geometry']]
```

```python
whole_df['Valid'] = True
whole_df['display'] = True
```

```python
import fiona
with fiona.Env(OSR_WKT_FORMAT="WKT2_2018"):
  whole_df.to_file(os.path.join(json_folder,course_id+'.json'), driver='GeoJSON')
```
